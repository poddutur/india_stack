```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
hrefs = list()

url = 'http://npcs.in/directory/directory/small-scale-industries/z,,13,0,64/index.html'
hrefs.append(url)
r = requests.get(url)
html = r.text
soup = BeautifulSoup(html,"html5lib")
all_urls = soup.find_all('a',attrs={"class":"b-pg-nav"})
for nurl in all_urls:
    if 'Next' in nurl.get('title'):
        nurl = nurl.get('href')
        break

while  ('/directory/directory/small-scale-industries/z,,' in str(nurl)) and (len(hrefs) < 100):
    nurl = str('http://npcs.in')+str(nurl)
    hrefs.append(nurl)
    r = requests.get(nurl)
    html = r.text
    soup = BeautifulSoup(html,"html5lib")
    all_urls = soup.find_all('a',attrs={"class":"b-pg-nav"})
    for nurl in all_urls:
        if 'Next' in nurl.get('title'):
            nurl = nurl.get('href')
            break
    print(len(hrefs),end='')
```

```python
print(len(hrefs))
small_scale_industries_data = list()
for rand in range(0,len(hrefs)):
    url = hrefs[rand]
    r = requests.get(url)
    html = r.text
    soup = BeautifulSoup(html,"html5lib")
    
    con_tags = soup.find_all('div',attrs={"class":"d-con"})
    t = con_tags[1]
    name = t.find('h2').text
    b_tags = t.find_all('b')
    try:
        mobile =''
        for b_tag in b_tags:
            if 'Mobile' in b_tag.text:
                mobile = b_tag.nextSibling.strip()
                mobile = mobile[-10:]
                break
    except:
        mobile =''

    desc = t.find('p',attrs={"class":"info"}).text

    zipcode = str(t.find('address').text)[-13:-7]

    try:
        website=''
        inside_tags = t.find_all('a')
        for inside_tag in inside_tags:
            if ('.com' in inside_tag.text) or ('.co.in' in inside_tag.text):
                website = inside_tag.text
                break
    except:
        website = ''

    try:
        telephone =''
        for b_tag in b_tags:
            if 'Tel' in b_tag.text:
                telephone = b_tag.nextSibling.strip()
                telephone = telephone[1:]
                telephone = telephone.strip()
                break
    except:
        telephone =''
    small_scale_industries_data.append({'Name': name,
                                       'Mobile':mobile,
                                       'Desc': desc,
                                       'Website':website,
                                       'Zipcode':zipcode,
                                       'Telephone':telephone})
    print(rand,end='')
small_scale_industries_data = pd.DataFrame(small_scale_industries_data)
```
